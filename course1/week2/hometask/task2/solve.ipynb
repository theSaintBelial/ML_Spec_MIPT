```python
%matplotlib inline

import numpy as np
from scipy.linalg import solve
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2))
```

```python
xold = np.arange(1, 15, 0.1)
yold = f(xold)
```

```python
plt.plot(xold, yold)
plt.show()
```

```python
x = np.array((1, 15))
b = f(x)

def getA(x, n):
    return np.array([[x0**n0 for n0 in range(n + 1)] for x0 in x])

A = getA(x, 1)
```

```python
X = solve(A, b)
```

```python
def g(x, X, n):
    y = 0
    for n0 in range(n + 1):
        y += X[n0]*(x**n0)
    return y

x = np.arange(1, 15+1, 1)
y = [g(x0, X, 1) for x0 in x]

plt.plot(x, y, xold, yold)
plt.show()
```

```python
x = np.array((1, 8, 15))

b = f(x)

A = getA(x, 2)

X = solve(A, b) # Ax = b
```

```python
x = np.arange(1, 15+1, 1)
y = [g(x0, X, 2) for x0 in x]

plt.plot(x, y, xold, yold)
plt.show()
```
