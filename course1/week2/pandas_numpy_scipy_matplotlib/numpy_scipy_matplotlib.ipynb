```python
%matplotlib inline

import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10, 2)
y = np.exp(-x / 3) + np.random.randn(len(x)) * 0.05

print(x[:5], y[:5], sep='\n------------------\n')
```

```python
# f = interpolate.interp1d(x, y, kind='linear')
f = interpolate.interp1d(x, y, kind='quadratic')
xnew = np.arange(0, 8, 0.1)
ynew = f(xnew)
```

```python
plt.plot(x, y, 'o', xnew, ynew)
plt.show()
```

```python
plt.plot(x, y, 'o')
plt.show()
```

```python
g = interpolate.interp1d(x, y, kind='linear')
```

```python
ynew = g(xnew)

plt.plot(xnew, ynew)
plt.show()
```
