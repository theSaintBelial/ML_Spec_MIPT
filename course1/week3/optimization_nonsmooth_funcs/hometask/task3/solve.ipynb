```python
import numpy as np

def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5*np.exp(-x / 2) 

def h(x):
    return int(f(x))
```

```python
%matplotlib inline

import matplotlib.pyplot as plt

x = np.arange(0, 30, .5)
y1 = f(x)
y2 = np.array([int(f(x_i)) for x_i in x])

plt.plot(x, y1, x, y2)
plt.show()
```

```python
x0 = 30

from scipy.optimize import fmin_bfgs

fmin_bfgs(h, x0)
```

```python

from scipy.optimize import minimize

minimize(h, x0, method='bfgs')
```

```python
from scipy.optimize import differential_evolution

bounds = [[1, 30]]

differential_evolution(h, bounds)
```
