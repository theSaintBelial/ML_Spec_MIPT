```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sts
```

```python
k = 1
tetta = 2

gammaRV = sts.gamma(a=k, scale=tetta)
sample = gammaRV.rvs(100)
sample.sum()
```

```python
x = np.linspace(0, 20, 100)

plt.hist(sample, density=True)
plt.plot(x, gammaRV.pdf(x))
plt.show()
```

```python
ex = k*tetta
dx = k*(tetta**2)
```

```python
ex, dx # theoretical
```

```python
def eex(sample):
    return sample.sum() / sample.shape[0]
```

```python
def func(n):
    count = 1000

    eexs = []
    for i in range(count):
        sample = gammaRV.rvs(n)
        eexs.append(eex(sample))

    exRV = sts.norm(loc=ex, scale=(dx / n)**.5)

    x = np.linspace(0, 10, 100)

    plt.hist(eexs, density=True)
    plt.plot(x, exRV.pdf(x))
    plt.show()
```

```python
func(5)
```

```python
func(10)
```

```python
func(50)
```
