```python
%matplotlib inline

import scipy.stats as sts
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu = 7
sigma = 4

normRV = sts.norm(loc=mu, scale=sigma)
```

```python
normRV.rvs(10)
```

```python
# x = np.arange(-2, 16, .1)
x = np.linspace(-2, 16, 100) # 3rd arg is the count, not step. More flexible
cdf = normRV.cdf(x)

plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
pdf = normRV.pdf(x)

plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
a = 5
b = 10

uniRV = sts.uniform(a, b-a)
uniRV.rvs(10)
```

```python
x = np.linspace(0, 15, 100)
cdf = uniRV.cdf(x)

plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
pdf = uniRV.pdf(x)

plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
p = 0.5

bernoulliRV = sts.bernoulli(p)

bernoulliRV.rvs(10)
```

```python
n = 10
p = 0.7

binomRV = sts.binom(n, p)
binomRV.rvs(10)
```

```python
x = np.linspace(0, 20, 21)

cdf = binomRV.cdf(x)

plt.plot(x, cdf)
plt.step(x, cdf)
```

```python
 pmf = binomRV.pmf(x)

 plt.plot(x, pmf, x, pmf, 'o')
```

```python
x = np.linspace(0, 50, 51)

for n in [20, 30]:
    for p in [0.2, 0.7]:
        rv = sts.binom(n, p)
        cdf = rv.cdf(x)
        plt.step(x, cdf, label="$N=%s, p=%s$" % (n,p))
plt.legend()
plt.title('CDF Binom')

plt.ylabel('$F(X)$')
plt.xlabel('$x$')
```

```python
x = np.linspace(0,45,46)
symbols = iter(['o', 's', '^', '+'])
for N in [20, 30]:
    for p in [0.2, 0.8]:
        rv = sts.binom(N, p)
        pmf = rv.pmf(x)
        plt.plot(x, pmf, next(symbols), label="$N=%s, p=%s$" % (N,p))
plt.legend()
plt.title("PMF (binomial)")

plt.ylabel('$P(X=x)$')
plt.xlabel('$x$')
```

```python
lmbda = 5

poissonRV = sts.poisson(lmbda)
poissonRV.rvs(10)
```

```python
x = np.linspace(0,30,31)
for l in [1, 5, 10, 15]:
    rv = sts.poisson(l)
    cdf = rv.cdf(x)
    plt.step(x, cdf, label="$\lambda=%s$" % l)
plt.legend()
plt.title("CDF (poisson)")

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```
