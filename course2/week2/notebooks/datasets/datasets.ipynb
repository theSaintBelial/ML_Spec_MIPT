```python
from sklearn import datasets
```

```python
%pylab inline
```

```python
circles_ds = datasets.make_circles()
```

```python
def plot_2d_ds(dataset, colors):
    pyplot.figure(figsize(5, 5))
    pyplot.scatter(list(map(lambda x: x[0], dataset[0])), list(map(lambda y: y[1], dataset[0])), c=dataset[1], cmap=colors)
```

```python
from matplotlib.colors import ListedColormap
```

```python
colors = ListedColormap(['red', 'blue'])

plot_2d_ds(circles_ds, colors)
```

```python
# with noise

noisy_circles_ds = datasets.make_circles(noise=0.1)
plot_2d_ds(noisy_circles_ds, colors)
```

```python
simplt_classification_problem_ds = datasets.make_classification(n_features=2, n_informative=1,
                                                                n_redundant=1, n_clusters_per_class=1,
                                                                random_state=1)
```

```python
plot_2d_ds(simplt_classification_problem_ds, colors)
```

```python
classification_problem_ds = datasets.make_classification(n_samples=10000, n_features=2, n_informative=2, n_classes=4,
                                                                n_redundant=0, n_clusters_per_class=1,
                                                                random_state=23)
colors = ListedColormap(['red', 'blue', 'yellow', 'black'])
plot_2d_ds(classification_problem_ds, colors)
```

```python
import pandas as pd
```

```python
iris_df = pd.DataFrame(iris.data)
```

```python
iris_df.columns = iris.feature_names
```

```python
iris_df['target'] = iris.target
iris_df['target'] = iris_df.target.apply(lambda index: iris.target_names[index])
```

```python
iris_df.head()
```

```python
iris_df[iris_df.target == 'setosa']['sepal length (cm)'].hist()
```

```python
 pyplot.figure(figsize(20, 20))
 
 index = 1
 for featureName in iris.feature_names:
     for targetName in iris.target_names:
         pyplot.subplot(4, 3, index)
         pyplot.hist(iris_df[iris_df.target == targetName][featureName])
         pyplot.title(targetName)
         pyplot.xlabel('cm')
         pyplot.ylabel(featureName[:-4])
         index += 1
```
